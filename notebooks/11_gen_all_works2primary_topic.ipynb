```python
works_by_topic_parquet_folder = "data/works_by_topic_parquet/"
all_works2primary_topic_parquet_folder = "data/all_works2primary_topic_parquet/"
os.makedirs(all_works2primary_topic_parquet_folder, exist_ok=True)
```

```python
topics = [topic[:-8] for topic in os.listdir(works_by_topic_parquet_folder)]
```

```python
len(topics)
```

```python
topic = topics[0]
all_works_df = read_parquet(works_by_topic_parquet_folder+topic, columns = ["id", "date", "primary_topic"])
```

```python
for topic in tqdm(topics[1:]):
    tmp_works_df = read_parquet(works_by_topic_parquet_folder+topic, columns = ["id", "date", "primary_topic"], quiet = True)
    all_works_df = pd.concat([all_works_df, tmp_works_df], ignore_index=True)
```

```python
all_works_df.to_parquet(all_works2primary_topic_parquet_folder+"all_works2primary_topic.parquet", compression = "brotli")
```
