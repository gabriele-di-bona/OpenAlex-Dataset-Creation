```python
import os
os.chdir("../") # resets notebook directory to repository root folder (DO ONLY ONCE!)
import time
import urllib.request
import xml.etree.ElementTree as ET

BASE_URL = "https://openalex.s3.amazonaws.com"
DOWNLOAD_DIR = "data/openalex-snapshot"
NS = {"s3": "http://s3.amazonaws.com/doc/2006-03-01/"}

def list_files(prefix):
    files = []
    marker = ""
    while True:
        url = f"{BASE_URL}/?prefix={prefix}&marker={marker}"
        with urllib.request.urlopen(url) as response:
            xml = response.read()
        root = ET.fromstring(xml)
        for contents in root.findall(".//s3:Contents", namespaces=NS):
            key = contents.find("s3:Key", namespaces=NS).text
            files.append(key)
        is_truncated = root.find("s3:IsTruncated", namespaces=NS)
        if is_truncated is not None and is_truncated.text == "true":
            marker = files[-1]
        else:
            break
    files_filtered = []
    for file in files:
        if "legacy-data" not in file and ( file.endswith(".gz") or file.endswith(".txt") ):
            files_filtered.append(file)
    return files_filtered

def download_file(key):
    dest_path = os.path.join(DOWNLOAD_DIR, key)
    if os.path.exists(dest_path):
        print(f"✔ Already downloaded: {key}")
        return
    os.makedirs(os.path.dirname(dest_path), exist_ok=True)
    url = f"{BASE_URL}/{key}"
    print(f"↓ Downloading {key}")
    try:
        urllib.request.urlretrieve(url, dest_path)
        time.sleep(0.5)
    except Exception as e:
        print(f"⚠ Failed to download {key}: {e}")

def main():
    start_ts = time.time()
    print(f"Start timestamp: {time.strftime('%Y-%m-%d %H:%M:%S', time.localtime(start_ts))}")
    print("🔍 Listing files in the datadump")
    files = list_files("")
    print(f"Found {len(files)} files")

    for file_key in files:
        download_file(file_key)

    end_ts = time.time()
    print(f"End timestamp: {time.strftime('%Y-%m-%d %H:%M:%S', time.localtime(end_ts))}")
    print(f"Elapsed time: {end_ts - start_ts:.2f} seconds ({time.strftime('%H:%M:%S', time.gmtime(end_ts - start_ts))})")

if __name__ == "__main__":
    main()
```
